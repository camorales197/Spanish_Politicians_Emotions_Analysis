```python
import twitter_utils
import pandas as pd
import plotly.express as px
import datetime

pd.set_option('display.max_colwidth', -1)
```

```python
authors_list = []
tweets_list = []
scores_list = []
dates_list = []

authors = ['sanchezcastejon', 'pablocasado_', 'PabloIglesias', 'Santi_ABASCAL', 'InesArrimadas']

for author in authors:
    tweets, dates, authors = twitter_utils.list_tweets(author, 1000)
    for t in tweets:
        text, sentiment = twitter_utils.sentiment_analyzer_scores(t)
        tweets_list.append(text)
        scores_list.append(sentiment)
    for d in dates:
        dates_list.append(d)
    for a in authors:
        authors_list.append(a)
        
```

```python
data = {'Author': authors_list,
        'Date': dates_list,
        'Tweet': tweets_list,
        'Score': scores_list
        }

df = pd.DataFrame (data, columns = ['Author', 'Date', 'Tweet', 'Score'])
```

```python
def truncate_date(input_date):
    return input_date.date()

df['Date'] = df.apply(lambda x: truncate_date(x['Date']),axis=1)
```

```python
df_group = df.groupby(['Date', 'Author']).mean()
df_group.reset_index(inplace=True)
```

```python
fig = px.line(df_group, x='Date', y='Score', color='Author', title="Análisis Temporal de Sentimiento Políticos España")
fig.show()
```
